```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense

penultimate_layer = model.get_layer('dense_fc').output
new_output = Dense(1, activation='sigmoid', name='continuous_score')(penultimate_layer)

continuous_score_model = Model(inputs=model.input, outputs=new_output)

for layer in continuous_score_model.layers:
    if layer.name in [l.name for l in model.layers] and 'output' not in layer.name:
        try:
            original_layer = model.get_layer(layer.name)
            original_weights = original_layer.get_weights()

            continuous_score_model.get_layer(layer.name).set_weights(original_weights)

        except Exception as e:
            print(f"Could not transfer weights for layer: {layer.name}. Error: {e}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
from matplotlib.backends.backend_pdf import PdfPages
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# --- Data Preparation ---
df_spleen = pd.DataFrame({
    'Age': age_data_spleen,
    'Score': spleen_scores,
    'Class': 'Spleen'
})
df_thymus = pd.DataFrame({
    'Age': age_data_thymus,
    'Score': thymus_scores,
    'Class': 'Thymus'
})

df_combined = pd.concat([df_spleen, df_thymus], ignore_index=True)
df_combined['Logged Age'] = np.log2(df_combined['Age'])

# --- Quadratic Regression Setup ---
def fit_quadratic_model(X, y):
    model = Pipeline([
        ('poly', PolynomialFeatures(degree=2)),
        ('linear', LinearRegression())
    ])
    model.fit(X, y)
    return model

# Create grids for plotting regression curves
min_age, max_age = df_combined['Age'].min(), df_combined['Age'].max()
min_logged, max_logged = df_combined['Logged Age'].min(), df_combined['Logged Age'].max()

age_grid = np.linspace(min_age, max_age, 100).reshape(-1, 1)
logged_age_grid = np.linspace(min_logged, max_logged, 100).reshape(-1, 1)

# Convert grids to DataFrames with the appropriate column names to match the training data
age_grid_df = pd.DataFrame(age_grid, columns=['Age'])
logged_age_grid_df = pd.DataFrame(logged_age_grid, columns=['Logged Age'])

# Fit quadratic models separately for each class and each type of age variable
models_age = {}
models_logged = {}
for cl in ['Spleen', 'Thymus']:
    df_cl = df_combined[df_combined['Class'] == cl]
    models_age[cl] = fit_quadratic_model(df_cl[['Age']], df_cl['Score'])
    models_logged[cl] = fit_quadratic_model(df_cl[['Logged Age']], df_cl['Score'])

# Get predictions on the grids using the DataFrames
predictions_age = {cl: models_age[cl].predict(age_grid_df) for cl in ['Spleen', 'Thymus']}
predictions_logged = {cl: models_logged[cl].predict(logged_age_grid_df) for cl in ['Spleen', 'Thymus']}

# --- Plotting Setup ---
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)
label_font = 18
title_font = 20

# Panel 0: Original Age vs Score
sns.scatterplot(
    data=df_combined, x='Age', y='Score', hue='Class',
    palette={'Spleen': 'red', 'Thymus': 'blue'},
    alpha=0.4, edgecolor=None, ax=axes[0], s=100
)
for cl, color in zip(['Spleen', 'Thymus'], ['red', 'blue']):
    axes[0].plot(age_grid, predictions_age[cl], color=color, lw=2, label=f'{cl} Quad Fit')
axes[0].set_xlabel("Age", fontsize=label_font)
axes[0].set_ylabel("Thymus-Spleen Model Score", fontsize=label_font)
axes[0].legend(title='Class', fontsize=18, title_fontsize=18, loc='upper right')

# Panel 1: Logged Age vs Score (default ticks)
sns.scatterplot(
    data=df_combined, x='Logged Age', y='Score', hue='Class',
    palette={'Spleen': 'red', 'Thymus': 'blue'},
    alpha=0.4, edgecolor=None, ax=axes[1], s=100,
    legend=False
)
for cl, color in zip(['Spleen', 'Thymus'], ['red', 'blue']):
    axes[1].plot(logged_age_grid, predictions_logged[cl], color=color, lw=2)
axes[1].set_xlabel("Age (log2)", fontsize=label_font)
axes[1].set_ylabel("")

# Panel 2: Logged Age vs Score (Custom Ticks)
sns.scatterplot(
    data=df_combined, x='Logged Age', y='Score', hue='Class',
    palette={'Spleen': 'red', 'Thymus': 'blue'},
    alpha=0.4, edgecolor=None, ax=axes[2], s=100,
    legend=False
)
unique_ages = np.unique(df_combined['Age'])

# Define the selected ages for vertical lines and ticks
selected_ages = np.array([1, 2, 3, 4, 8, 24, 36, 134, 290, 400])
logged_selected = np.log2(selected_ages)
axes[2].set_xticks(logged_selected)
axes[2].set_xticklabels(selected_ages.astype(int))
for x in logged_selected:
    axes[2].axvline(x=x, color='gray',  linewidth=0.5)

axes[2].set_xlabel("Age (log2)", fontsize=label_font)
axes[2].set_ylabel("")

for cl, color in zip(['Spleen', 'Thymus'], ['red', 'blue']):
    axes[2].plot(logged_age_grid, predictions_logged[cl], color=color, lw=2)
axes[2].set_xlabel("Age (log2)", fontsize=label_font)
axes[2].set_ylabel("")

plt.tight_layout()


# Export the plot as a PDF and the DataFrame as CSV
base_export_dir = 'CNNContinousScale'
os.makedirs(base_export_dir, exist_ok=True)
pdf_path = os.path.join(base_export_dir, 'TrainingData_combined_plot.pdf')
with PdfPages(pdf_path) as pdf:
    pdf.savefig(fig)
print(f"Saved the combined plot to {pdf_path}")

csv_path = os.path.join(base_export_dir, 'TrainingData_combined_plot.csv')
df_combined.to_csv(csv_path, index=False)
print(f"Saved the combined DataFrame to {csv_path}")

plt.show()
```

```python
# gradcam_module.py

# Imports
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy.ndimage import gaussian_filter, zoom
from tensorflow.keras.models import load_model



def make_gradcam_heatmap(img_array, model, conv_layer, age_array, timerpos_array=None, optimal_threshold=None, pred_index=1):
    """
    Generates a GradCAM heatmap for a specific class index based on an optimal threshold logic.
    Focuses on the specified class if its probability is above a certain threshold.

    Args:
        img_array (np.array): Input image for model prediction.
        model (tf.keras.Model): Trained Keras model.
        conv_layer (str): Name of the last convolutional layer to use for GradCAM.
        age_array (np.ndarray): Array containing age data reshaped to (batch_size, 1).
        timerpos_array (np.ndarray, optional): Timer-positive numeric array of shape (batch_size, D).
        optimal_threshold (float, optional): Threshold to decide if focusing on the pred_index.
        pred_index (int): Index of the class (default 1 for WT in [KO, WT]).
    """
    grad_model = tf.keras.models.Model(
        inputs=model.inputs,
        outputs=[model.get_layer(conv_layer).output, model.output]
    )
    
    inputs_to_model = [img_array]  # Initialize with image array
    if timerpos_array is not None:
        inputs_to_model.append(timerpos_array)
    if age_array is not None:
        inputs_to_model.append(age_array)

    with tf.GradientTape() as tape:
        conv_layer_output, preds = grad_model(inputs_to_model)

        prob_class_of_interest = preds[0][pred_index]
        
        if optimal_threshold is not None:
            if prob_class_of_interest >= optimal_threshold:
                focus_index = pred_index
            else:
                focus_index = 1 - pred_index
        else:
                    focus_index = pred_index
                    
        class_channel = preds[:, focus_index]

    grads = tape.gradient(class_channel, conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    conv_layer_output = conv_layer_output[0]
    heatmap = conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)

    heatmap = tf.maximum(heatmap, 0)  # Apply ReLU
    heatmap /= (tf.reduce_max(heatmap) + 1e-10)  # Normalize to [0, 1]

    return heatmap.numpy()



def smooth_heatmap(heatmap, sigma=2):
    """Normalizes and applies Gaussian smoothing to the heatmap."""
    # Apply Gaussian smoothing
    smoothed_heatmap = gaussian_filter(heatmap, sigma=sigma)
    
    # Calculate mean and standard deviation
    mean_val = np.mean(smoothed_heatmap)
    std_val = np.std(smoothed_heatmap)
    
    # Normalize the heatmap based on the standard deviation
    if std_val > 1e-3:
        normalized_heatmap = (smoothed_heatmap - mean_val) / std_val
    else:
        normalized_heatmap = smoothed_heatmap - mean_val
    
    return normalized_heatmap

def generate_density(images):
    """Calculates the average pixel intensity across all images, normalized to sum to 1."""
    density = np.mean(images, axis=0)
    return density / np.sum(density)

def generate_aggregated_heatmap(images, age_data, model, last_conv_layer_name, timerpositives=None, threshold_value=0.0, optimal_threshold=None, default_pred_index=0):
    """
    Generates an aggregated GradCAM heatmap for a set of images, optionally including
    timer-positive data and age data for each image.

    Args:
        images (np.ndarray): Array of shape (N, H, W, C), where N is # of images.
        age_data (np.ndarray): Numeric data of shape (N,) corresponding to age for each image.
        model (tf.keras.Model): Trained Keras model.
        last_conv_layer_name (str): Name of the final convolutional layer.
        timerpositives (np.ndarray, optional): Timer data.
        threshold_value (float, optional): Threshold for zeroing out low heatmap values.
        optimal_threshold (float, optional): Threshold to decide class index focus.
        default_pred_index (int): Default class index for focus if no threshold is used.
    """
    aggregated_heatmap = np.zeros((images.shape[1], images.shape[2]), dtype=np.float32)
    epsilon = 1e-10
    
    for i in range(len(images)):
        img = images[i:i+1]  # Slice to keep the batch dimension
        timerpos = timerpositives[i:i+1] if timerpositives is not None else None
        age = np.array([age_data[i]])  # Reshape age to (1, 1) if necessary
        
        heatmap = make_gradcam_heatmap(img, model, last_conv_layer_name, age, timerpos, optimal_threshold, default_pred_index)
        heatmap = np.expand_dims(zoom(heatmap, (images.shape[1] / heatmap.shape[0], images.shape[2] / heatmap.shape[1]), order=1), axis=0)
        
        if np.sum(heatmap) > 0:
            heatmap /= np.sum(heatmap + epsilon)
        
        aggregated_heatmap += heatmap.squeeze()  # Remove batch dimension after processing

    if np.max(aggregated_heatmap) > 0:
        aggregated_heatmap /= np.max(aggregated_heatmap + epsilon)

    return aggregated_heatmap

def visualize_heatmaps(aggregated_heatmap, smoothed_heatmap, difference_heatmap, title_suffix):
    """Visualizes raw and smoothed heatmaps along with their differences.
    
    Parameters:
    - aggregated_heatmap: numpy array of the raw GradCAM heatmap.
    - smoothed_heatmap: numpy array of the Gaussian-smoothed GradCAM heatmap.
    - difference_heatmap: numpy array showing the difference between two heatmaps.
    - title_suffix: str, suffix to add to the plot titles to indicate the dataset or condition.
    """
    fig, ax = plt.subplots(1, 3, figsize=(15, 5))
    cmap = plt.get_cmap('viridis')
    diff_cmap = plt.get_cmap('coolwarm')

    # Display the aggregated heatmap
    ax[0].imshow(aggregated_heatmap, cmap=cmap, origin='lower')
    ax[0].set_title(f"Raw GradCAM {title_suffix}")
    ax[0].axis('off')

    # Display the smoothed heatmap
    ax[1].imshow(smoothed_heatmap, cmap=cmap, origin='lower')
    ax[1].set_title(f"Smoothed GradCAM {title_suffix}")
    ax[1].axis('off')

    # Display the difference heatmap
    ax[2].imshow(difference_heatmap, cmap=diff_cmap, origin='lower')
    ax[2].set_title(f"Difference Heatmap {title_suffix}")
    ax[2].axis('off')

    plt.tight_layout()
    plt.show()

def find_optimal_threshold(fpr, tpr, thresholds, factor=0.5):
    """Find an optimal threshold for a group identification."""
    gmeans = np.sqrt(tpr * (1-fpr))
    index = np.argmax(gmeans)
    return thresholds[index]

```
